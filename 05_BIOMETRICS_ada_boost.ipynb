```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
ada_clf = AdaBoostClassifier(DecisionTreeClassifier(random_state = 42), 
                             algorithm="SAMME.R", 
                             random_state = 42)
```

```python
learning_pts = []
learning_threshold = 0
for x in range(0,10):
    learning_threshold += 0.1
    learning_pts.append(learning_threshold)
print(learning_pts)
```

```python
param_grid = [
#     {'n_estimators': [i for i in range(20,40,2)], {'base_estimator__max_depth': 2, 'learning_rate': 0.6, 'n_estimators': 22}
    {'n_estimators': [i for i in range(20,25)], 
    'learning_rate': learning_pts,
    "base_estimator__max_depth" : [i for i in range(1,10)]} 
#       {'n_estimators': [i for i in range(1,10)], 'learning_rate': learning_pts}
]
```

```python
grid_ada_clf = GridSearchCV(ada_clf, param_grid, cv=5, verbose=10)
```

```python
grid_ada_clf.fit(X_train_std, y_train)
```
